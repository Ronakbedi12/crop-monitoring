```python
grouped=df.groupby('label')
grouped[['temperature']].mean()
```

```python
grouped[['N']].mean()
```

```python
grouping=grouped['temperature'].agg(['min', 'max', 'mean'])
grouping.head()
```

```python
sns.barplot(grouping)
plt.show()
```

```python
grouping.reset_index(inplace=True)
x=list(grouping['label'])
y=list(grouping['mean'])
plt.figure(figsize=(14,7))
plt.bar(x,y,ec='k',color='maroon')
plt.title('Average temperature')
plt.xlabel('Label')
plt.ylabel('min')
plt.xticks(rotation=90)
plt.show()
```

```python
grouping.reset_index(inplace=True)
x=list(grouping['label'])
y=list(grouping['min'])
plt.figure(figsize=(14,7))
plt.bar(x,y,ec='k',color='maroon')
plt.title('min temperature')
plt.xlabel('Label')
plt.ylabel('min')
plt.xticks(rotation=90)
plt.show()
```

```python
grouping.reset_index(inplace=True)
x=list(grouping['label'])
y=list(grouping['max'])
plt.figure(figsize=(14,7))
plt.bar(x,y,ec='k',color='green')
plt.title('Max temperature')
plt.xlabel('Label')
plt.ylabel('max')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.lineplot(grouping)
plt.xlabel('temperature')
plt.ylabel('production')
plt.title('temperature vs production')
plt.show()
```

```python
## to normalize the outliers
from sklearn.preprocessing import StandardScaler,MinMaxScaler
std=StandardScaler()
x_transformed=std.fit_transform(x)
```

```python
print("Total no. of clay samples: ",x_transformed.shape[0])
```

Training and testing spliting

```python
train=x_transformed[:2100]
test=x_transformed[2100:]
```

```python
## unsupervised 
from sklearn.cluster import KMeans
obj=KMeans()
```

```python
obj.fit(train)
```

```python
print("Training score:",round(obj.score(train)))
print("Testing score:",round(obj.score(test)*100,2))
```
